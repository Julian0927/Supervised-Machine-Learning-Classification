# ***Gradient Boosting for classification with Python***

$ \ $

$(1)$ In this notebook, you will learn Gradient Boosting for classification; AdaBoost is a particular case of Gradient Boosting and an additive model where we add weak learners to minimize the loss function. 

$ \ $

$(2)$ This lab will focus on <a href="https://xgboost.readthedocs.io/en/stable/?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMML241ENSkillsNetwork31576874-2022-01-01">XGBoost</a>, an open-source software library that provides a regularizing gradient boosting framework.  XGBoost can use different types of weak learners called boosters for classification and regression. We will focus on trees for classification.

$ \ $

$(3)$ Unlike Bagging and Random Forest, Gradient Boosting can cause overfitting. As a result, Gradient Boosting requires Hyperparameter tuning, taking more time to train.

$ \ $

$(4)$ One advantage of Gradient Boosting is that each classifier is smaller, so predictions are faster in Gradient Boosting.
AdaBoost is a subclass of Gradient Boosting; one weakness of AdaBoost is that misclassified samples cause overfitting; Gradient Boosting uses different loss functions, reducing this effect.

$ \ $

$(5)$ The following table show's the Average accuracy and standard deviation for:

* Random Forest (RF),

* gradient boosting (GB)

* XGBoost,

both using the default (D) and (T) tuned parameter settings, we see that XGBoost does best, followed by GB outperforming the other methods.

## ***Some important functions***

$ \ $

$(1)$ This function will calculate the accuracy of the training and testing data given a model.

```python
def get_accuracy(X_train, X_test, y_train, y_test, model):

  # predecimos un conjunto de datos con la data de prueba y calculamos la precision entre y_test, y_pred
  y_pred_test = model.predict(X_test)
  accuracy_test = metrics.accuracy_score(y_test, y_pred_test)

  # predecimos un conjunto de datos con la data de  entrenamiento y calculamos la precision entre y_test, y_pred
  y_pred_train = model.predict(X_train)
  accuracy_train = metrics.accuracy_score(y_train, y_pred_train)

  # la funcion devuelve 
  return  {"test Accuracy": accuracy_test, "train Accuracy": accuracy_train}
```

$ \ $

$(2)$ This function calculates the average accuracy of differnt learning rates on training and test data.

```python
def get_accuracy_boost(X, y, title, times = 20, xlabel = 'Number Estimators', Learning_rate_ = [0.2, 0.4, 0.6, 1], n_est = 100):

    # creamos 4 tipos de lineas para nuestros plots. Esto se hace por medio de una lista con la siguiente descripcion.
    lines_array = ['solid', '--', '-.', ':']

    # creamos una lista con los numeros 2, 4, 6,..., n_est
    N_estimators = [n*2 for n in range(1, n_est//2)]
    
    # creamos N - matrices nulas de tamao mxn
    N = times
    m = len(Learning_rate_)
    n = len(N_estimators)
    train_acc = np.zeros((N, m, n))

    # creamos N - matrices nulas de tamao mxn
    test_acc = np.zeros((N, m, n))

    # tqdm nos ayuda a crear barras de progreso para el procesamiento de datos, el entrenamiento de modelos de aprendizaje automtico    
    # se realiza una iteracion sobre [0,1,...,N-1]
    for n in tqdm(range(N)):

        # separamos nuestra data en data de entrenamiento (70%) y data de prueba (30%)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

        # realizamos una iteracion sobre [2, 4, 6,..., n_est]
        for n_estimators in N_estimators:

            #  si Learning_rate_ = [0.2, 0.4, 0.6, 1], entonces realizamos una iteracion con indices de esta lista
            for i, lr in enumerate(Learning_rate_):

                # definimos la variable j
                j = (n_estimators//2)-1

                # creamos el objeto XGBClassifier con los parametros establecidos
                model = XGBClassifier(objective = "binary:logistic", learning_rate = lr, n_estimators = n_estimators, eval_metric = 'mlogloss')

                # entrenamos nuestro modelo con la data de entrenamiento
                model.fit(X_train, y_train)

                # obtenemos la precision del modelo 
                Accuracy = get_accuracy(X_train, X_test, y_train, y_test, model)

                # actualizamos la n-esima matriz de train_acc
                train_acc[n, i, j] = Accuracy['train Accuracy']

                # actualizamos la n-esima matriz de test_acc
                test_acc[n, i, j] = Accuracy['test Accuracy']   

    # creamos una imagen donde describiremos varios subplots
    fig, ax1 = plt.subplots()

    # calculamos el promedio de las columnas de test_acc
    mean_test = test_acc.mean(axis = 0)

    # calculamos el promedio de las columnas de train_acc
    mean_train = train_acc.mean(axis = 0)

    # creamos ejes gemelos que comparten el eje x.
    ax2 = ax1.twinx()

    # Learning_rate_ = [0.2, 0.4, 0.6, 1]
    # lines_array = ['solid', '--', '-.', ':']
    for i, (lr, line) in enumerate(zip(Learning_rate_, lines_array)): 

        # realizamos una grafica sobre el primer plot usando mean_train[i, :]
        ax1.plot(mean_train[i, :], linestyle = line, color = 'b', label = "Learning rate " + str(lr))

        # realizamos una grafica sobre el segundo plot usando mean_train[i, :]
        ax2.plot(mean_test[i, :], linestyle = line, color = 'r', label = str(lr))

    # establecemos la etiqueta del titulo del eje y del plot 1 con color rojo
    ax1.set_ylabel('Training accuracy', color = 'b')

    # establecemos la leyenda del plot 1
    ax1.legend() 

    # establecemos la etiqueta del titulo del eje y del plot 2 con color rojo
    ax2.set_ylabel('Testing accuracy', color = 'r')

    # establecemos la leyenda del plot 2
    ax2.legend()

    # establecemos la etiqueta del titulo del eje x
    ax1.set_xlabel(xlabel)

    # la funcion devuelve el imagen creada anteriormente
    return plt.show()
```

## ***Early Stopping***

Early Stopping is a way to prevent overfitting; we stop gradient descent based on a specified number of epochs over which no improvement is observed using the evaluation set .

## ***Parameters for Trees***

$ \ $

$(1)$ We can also change tree parameters, consider our first example.

```python
# modificamos los parametros de nuestro modelo 
model = XGBClassifier(objective = 'binary:logistic',
                      learning_rate = 0.1,
                      n_estimators = 10,
                      eval_metric = 'mlogloss')

# entrenamos nuestro modelo
model.fit(X_train, y_train)
```

```python
get_accuracy(X_train, X_test, y_train, y_test,  model)
```

$ \ $

$(2)$ The parameter `max_depth` is the maximum depth of a tree, and increasing this value will make overfitting more likely. Do decreasing max depth will decrease overfitting.

```python
# creamos el modelo con una maxima profundidad del arbol de 3
model = XGBClassifier(objective = 'binary:logistic',
                      learning_rate = 0.1,
                      n_estimators = 10,
                      eval_metric = 'mlogloss',
                      max_depth = 3)

# entrenamos nuestro modelo
model.fit(X_train, y_train)
```

```python
get_accuracy(X_train, X_test, y_train, y_test,  model)
```

$ \ $

$(3)$ The parameter `min_child_weight` is essentially the minimum number of instances needed to be in each node,  larger values decrease overfitting. The default is one and if we set it equal to four, then we get better results on the training data.


```python
# creamos el modelo con minimamente 4 numero de instancias en cada nodo
model = XGBClassifier(objective = 'binary:logistic',
                      learning_rate = 0.1,
                      n_estimators = 10,
                      eval_metric = 'mlogloss',
                      min_child_weight = 4)

# entrenamos nuestro modelo
model.fit(X_train, y_train)
```

```python
# creamos de nuevo el modelo clasificador XGB con los parametros mostrados
model = XGBClassifier(objective = 'binary:logistic',
                      learning_rate = 0.1,
                      n_estimators = 10,
                      eval_metric = 'mlogloss',
                      gamma = 1,
                      reg_lambda = 2,
                      alpha = 1)

# entrenamos nuestro modelo
model.fit(X_train, y_train)
```

```python
get_accuracy(X_train, X_test, y_train, y_test,  model)
```

## ***Cancer Data Example with GridSearchCV***

$ \ $

The example is based on a dataset that is publicly available from the UCI Machine Learning Repository [Cancer data](http://mlearn.ics.uci.edu/MLRepository.html?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMML241ENSkillsNetwork31576874-2022-01-01). The dataset consists of several hundred human cell sample records, each of which contains the values of a set of cell characteristics. 

$ \ $

$(1)$ Let's load the dataset.

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/cell_samples.csv")
df.head()
```

$ \ $

$(2)$ Now let's remove rows that have a ***?*** in the <code>BareNuc</code> column.


```python
df = df[df["BareNuc"] != "?"]
df
```

$ \ $

$(3)$ We obtain the features and class labels.

```python
# describimos los rasgos de X
X =  df[['Clump', 'UnifSize', 'UnifShape', 'MargAdh', 'SingEpiSize', 'BareNuc', 'BlandChrom', 'NormNucl', 'Mit']]
X = X.astype('int64')
```

```python
# creamos la data objetivo
y = df[['Class']]
y.value_counts()
```

```python
# creamos un codificador de etiquetas para transformar los datos de la columna 3 de X, en datos numericos
le_Chol = preprocessing.LabelEncoder()

# entrenamos nuestro modelo con los valores de la columna 3 de X
le_Chol.fit([2, 4])

# transormamos los datos de la columna 3 de X en datos numericos con el modelo codificador de etiquetas creado anteriormente
y = le_Chol.transform(y)
```

```python
model = XGBClassifier(objective = 'binary:logistic',
                      eval_metric = 'mlogloss')
```

$ \ $

$(6)$ We create the Grid Search object and fit it.

```python
# creamos los parametros del modelo GridSearchCV
param_grid = {'learning_rate': [0.1*(n+1) for n in range(5)], 'n_estimators' : [2*n+1 for n in range(5)]}

# creamos el modelo GridSearchCV con los parametros establecidos
search = GridSearchCV(estimator = model,
                      param_grid = param_grid,
                      scoring = "neg_log_loss")

# entrenamos nuestro modelo
search.fit(X_train, y_train)
```

$ \ $

$(7)$ The best parameter score are.


```python
search.best_score_ 
```

$ \ $

$(8)$ The best parameter values are.

```python
search.best_params_
```

$ \ $

$(9)$ We can calculate accuracy on the test data using the test data.

```python
get_accuracy(X_train, X_test, y_train, y_test, search.best_estimator_)
```

We can see the best accuracy score of the searched parameters was ~96%.

Imagine that you are a medical researcher compiling data for a study. You have collected data about a set of patients, all of whom suffered from the same illness. During their course of treatment, each patient responded to one of 5 medications:

* Drug A, 

* Drug B, 

* Drug C,

* Drug x,

Part of your job is to build a model to find out which drug might be appropriate for a future patient with the same illness. The $\color{yellow}{\text{features}}$ of this dataset are:

* Age, 

* Sex, 

* Blood Pressure, 

* the Cholesterol of the patients,

and the $\color{orange}{\text{target}}$ is 

* the drug that each patient responded to.

It is a sample of multiclass classifier, and you can use the training part of the dataset to:

* build a decision tree,

* predict the class of a unknown patient, or to prescribe a drug to a new patient. 

You will use GridSearchCV and Stacking Classifiers  to find the best results.

$ \ $

$(1)$ Let's load the dataset.



```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/drug200.csv", delimiter=",")
df.head()
```

$ \ $ 

$(2)$ Let's create an array to get the X and y for our dataset.

```python
data = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K', 'Drug']].values
data
```

$ \ $

$(3)$ Now let's use a <code>LabelEncoder</code> to turn categorical features into numerical.


```python
# creamos un codificador de etiquetas para transformar los datos de la columna 1 de X, en datos numericos
le_sex = preprocessing.LabelEncoder()

# entrenamos nuestro modelo con los valores de la columna 1 de X
le_sex.fit(['F','M'])

# transformamos los datos de la columna 1 de X en datos numericos con el modelo codificador de etiquetas creado anteriormente
data[:, 1] = le_sex.transform(data[:, 1]) 
```

```python
# creamos un codificador de etiquetas para transformar los datos de la columna 2 de X, en datos numericos
le_BP = preprocessing.LabelEncoder()

# entrenamos nuestro modelo con los valores de la columna 2 de X
le_BP.fit(['LOW', 'NORMAL', 'HIGH'])

# transformamos los datos de la columna 2 de X en datos numericos con el modelo codificador de etiquetas creado anteriormente
data[:,2] = le_BP.transform(data[:,2])
```

```python
# creamos un codificador de etiquetas para transformar los datos de la columna 3 de X, en datos numericos
le_Chol = preprocessing.LabelEncoder()

# entrenamos nuestro modelo con los valores de la columna 3 de X
le_Chol.fit(['NORMAL', 'HIGH'])

# transformamos los datos de la columna 3 de X en datos numericos con el modelo codificador de etiquetas creado anteriormente
data[:, 3] = le_Chol.transform(data[:, 3]) 
```

```python
# creamos un codificador de etiquetas para transformar los datos de la columna 3 de X, en datos numericos
le_dr = preprocessing.LabelEncoder()

# entrenamos nuestro modelo con los valores de la columna 3 de X
le_dr.fit(['drugA', 'drugB', 'drugC', 'drugX', 'drugY'])

# transformamos los datos de la columna 3 de X en datos numericos con el modelo codificador de etiquetas creado anteriormente
data[:, 5] = le_dr.transform(data[:, 5]) 
```

$ \ $ 

$(4)$ Let's create an array to get the X and y for our dataset.

```python
data = pd.DataFrame(data, columns = ["Age", "Sex", "BP",	"Cholesterol", "Na_to_K", "Drug"])
X = data[["Age", "Sex", "BP",	"Cholesterol", "Na_to_K"]].values
y = data["Drug"].astype('int')
```

$ \ $

$(5)$ Split the data into training and testing data with a $80/20$ split.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 4)
```

```python
print('Train set:', X_train.shape,  y_train.shape)
print('Test set:', X_test.shape,  y_test.shape)
```

$ \ $

$(6)$ We can use GridSearch for Exhaustive search over specified parameter values.

```python
# creamos los parametros
param_grid = {'learning_rate': [0.1*(n+1) for n in range(2)], 'n_estimators' : [2*n+1 for n in range(2)] }

# creamos un modelo clasificador XGB
model = XGBClassifier(objective = 'binary:logistic',
                      eval_metric = 'mlogloss')

# creamos un modelo GridSearchCV con los parametros establecidos
search = GridSearchCV(estimator = model,
                      param_grid = param_grid,
                      scoring = "neg_log_loss")

# entrenamos nuestro modelo GridSearchCV
search.fit(X_train, y_train)
```

```python
search.best_score_
```

```python
get_accuracy(X_train, X_test, y_train, y_test, search.best_estimator_)
```
